#Define Model (RESNET 34)

```
help(ImageDataBunch.from_folder)
```

```
bs = 64
tfms = get_transforms(do_flip=False)
data = ImageDataBunch.from_folder(data_dir, size=224, bs = bs, ds_tfms = tfms
                                 ).normalize(imagenet_stats)
```

```
data.show_batch(rows=4, figsize=(7,6))
```

```
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```
learn.fit_one_cycle(10)
```

```
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```
interp.plot_top_losses(9, figsize=(15,11))
```

```
learn.save(model_dir + 'kanye_not_kanye-32')
```

#Define Model (RESNET 50)

```
learn = cnn_learner(data, models.resnet50, metrics=error_rate)
```

```
learn.lr_find()
learn.recorder.plot()
```

```
learn.fit_one_cycle(10)
#learn.fit_one_cycle(10, max_lr=slice(1e-2))
```

```
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```
interp.plot_top_losses(9, figsize=(15,11))
```

```
learn.save(model_dir + 'kanye_not_kanye-50')
```
